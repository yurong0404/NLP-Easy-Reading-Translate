```python
from bs4 import BeautifulSoup
from collections import defaultdict
import requests
```

```python
def extract(soup):
    f = open('5000_voc_Oxford.txt','w')
    for big_block in soup.select('.responsive_entry_center'):
        for entry in big_block.select('.information-content-reduced-width'):
            for words in entry.select('.top-g'):
                count = 0
                for word in words.select('span'):
                    if (count%3 == 0):
                        f.write(word.text+'\t')
                        print(word.text)
                    elif (count%3 == 1):
                        f.write(word.text+'\n')
                    count += 1
    f.close()
                
        #        for sent in block.select('.extraexamps .eg'):
#            extra_sents.append(sent.text)
```

```python
def start(url,head):
    headers = { 'User-Agent': 'Mozilla/5.0 (Windows NT 6.0; WOW64; rv:24.0) Gecko/20100101 Firefox/24.0' }
    source_code = requests.get(url , headers=headers).content
    soup = BeautifulSoup(source_code, 'html.parser')
    extract(soup)
```
